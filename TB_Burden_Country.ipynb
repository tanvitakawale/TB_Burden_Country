QUESTIONS

###Descriptive Analysis

1)What are the top 5 countries with the highest estimated total population?

```python
total_population= tb.groupby('Country or territory name')['Estimated total population number'].sum()
top_5_countries=total_population.sort_values(ascending=False).head(5)
print("Highest estimated total population by",top_5_countries)

plt.figure(figsize=(10,6))
plt.bar(top_5_countries.index,top_5_countries.values,color=['skyblue','pink','yellow','purple','lightgreen'],width=0.8,label=top_5_countries.index)
plt.title('Top 5 countries with highest estimated total population')
plt.xlabel('Country Name')
plt.ylabel('Estimated Total population number')
plt.xticks()
plt.show()
```

###Temporal Analysis

2)Calculate the highest estimated total population by year?

```python
total_population= tb.groupby('Year')['Estimated total population number'].sum()
avg_total=total_population.sort_values(ascending=True).head(5)
print("Highest estimated total population by",avg_total)

plt.figure(figsize=(10,6))
plt.bar(avg_total.index,avg_total.values,color=['skyblue','pink','yellow','purple','lightgreen'],width=0.8,label=avg_total.index)
plt.title('Top 5 countries with highest estimated total population')
plt.xlabel('Year')
plt.ylabel('Estimated Total population number')
plt.xticks()
plt.show()
```

3)Calculated the average Estimated prevalence of TB (all forms) per 100 000 population by year?

```python
avg= tb.groupby('Year')['Estimated prevalence of TB (all forms) per 100 000 population'].sum()
avg_est=avg.sort_values(ascending=False).head(5)
print("Estimated prevalence of TB (all forms) per 100 000 population by",avg_est)

plt.figure(figsize=(11,8))
plt.bar(avg_est.index,avg_est.values,color=['skyblue','pink','yellow','purple','brown'],width=0.8,label=avg_est.index)
plt.title('Estimated prevalence of TB (all forms) per 100 000 population')
plt.xlabel('Year')
plt.ylabel('Estimated prevalence of TB (all forms) per 100 000 population')
plt.xticks()
plt.show()
```

4)How does Case detection rate vary by year?

```python
total_casedetection_rate= tb.groupby('Country or territory name')['Case detection rate (all forms), percent'].sum()
total_cases=total_casedetection_rate.sort_values(ascending=True).head(3)
print("Total case detection rate by",total_cases)

plt.figure(figsize=(10,6))
plt.bar(total_cases.index,total_cases.values,color=['skyblue','pink','yellow','purple','lightgreen'],width=0.8,label=total_cases.index)
plt.title('Total case detection rate by region')
plt.xlabel('Country or territory name')
plt.ylabel('Case detection rate (all forms), percent')
plt.xticks()
plt.show()
```

5)What is trend of the estimated prevalence of TB (all forms),low bound over the years for a specific country?

```python
country_name="Angola"
country_data=tb[tb['Country or territory name']==country_name]
print(country_name)

plt.figure(figsize=(12,6))
plt.plot(country_data['Year'],country_data['Estimated prevalence of TB (all forms) per 100 000 population, low bound'],marker='o')
plt.title('Trend of estimated prevelance of TB(all forms).low bound in {Country_name}')
plt.xlabel('Year')
plt.ylabel('Estimated prevalence of TB (all forms) per 100 000 population, low bound')
plt.grid(True)
plt.show()
```

###Country Analysis

11)Calculate bottom 2 Estimated prevalence of TB (all forms) per 100 000 population, high bound by Country name?

```python
total_population= tb.groupby('Country or territory name')['Estimated prevalence of TB (all forms) per 100 000 population, high bound'].mean()
top_5_countries=total_population.sort_values(ascending=False).tail(2)
print("lowest Estimated prevalence of TB (all forms) per 100 000 population, high bound by",top_5_countries)

plt.figure(figsize=(10,6))
plt.bar(top_5_countries.index,top_5_countries.values,color=['skyblue','pink','yellow','purple','lightgreen'],width=0.8,label=top_5_countries.index)
plt.title('Bottom 2 estimated prevalence of TB(all forms) per 100000 population,high bound')
plt.xlabel('Country Name')
plt.ylabel('Estimated prevalence of TB (all forms) per 100 000 population, high bound')
plt.xticks()
plt.show()
```

###Methodology Analysis

12)Calculate maximum Estimated prevalence of TB (all forms),low bound by Method to derive prevalence estimates?

```python
total_population= tb.groupby('Method to derive prevalence estimates')['Estimated prevalence of TB (all forms), low bound'].max()
print("Estimated prevalence of TB (all forms), low bound by",total_population)

plt.figure(figsize=(10,6))
plt.bar(total_population.index,total_population.values,color=['skyblue','pink','yellow','purple','lightgreen'],width=0.8,label=total_population.index)
plt.title('Distribution of Estimated prevalence of TB (all forms), low bound')
plt.xlabel('Method to derive prevalence estimates')
plt.ylabel('Estimated prevalence of TB (all forms), low bound')
plt.xticks()
plt.show()
```

13)How does the method to derive prevalence estimates affect the estimated mortality of TB cases(all forms,excluding HIV) per 100,000 population,high bound?

```python
total_population= tb.groupby('Method to derive prevalence estimates')['Estimated mortality of TB cases (all forms, excluding HIV), per 100 000 population, high bound'].mean()
print("Estimated mortality of TB cases (all forms, excluding HIV), per 100 000 population, high bound by",total_population)

plt.figure(figsize=(12,9))
plt.bar(total_population.index,total_population.values,color=['skyblue','pink','yellow','purple','lightgreen'],width=0.8,label=total_population.index)
plt.title('Distribution of Estimated mortality of TB cases')
plt.xlabel('Method to derive prevalence estimates')
plt.ylabel('Estimated mortality of TB cases (all forms, excluding HIV), per 100 000 population, high bound')
plt.xticks()
plt.show()
```

###Year Wise Analysis

14)Calculate bottom 4 Estimated prevalence of TB (all forms) per 100 000 population, high bound by Year?
